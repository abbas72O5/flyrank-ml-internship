## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# CODE CELL for Section 3
# 1. Compare the age of 'Up' pages vs 'Down' pages
age_down = df[df['trend_direction'] == 'down']['days_since_last_update'].median()
age_up = df[df['trend_direction'] == 'up']['days_since_last_update'].median()

# 2. Check the position loss for decaying pages
pos_down = df[df['trend_direction'] == 'down']['avg_position'].mean()
pos_up = df[df['trend_direction'] == 'up']['avg_position'].mean()

print(f"Median age of declining pages: {age_down} days")
print(f"Median age of growing pages: {age_up} days")
print(f"Mean position of declining pages: {pos_down:.1f} vs Growing: {pos_up:.1f}")
```

The Data: Declining pages are younger (20 days) than growing pages (22 days).

The Insight: Content "Decay" isn't happening to old, dusty pages; it's happening to new content.

The Discovery: This suggests a "Freshness Gap." New pages are being published, getting a small initial boost, and then immediately failing.

The ML Opportunity: A simple rule like "Refresh pages older than 6 months" would be totally useless here because the problem is happening to 20-day-old pages! You need ML to find the subtle signals (CTR, Position wiggles) because the "Age" signal is counter-intuitive.
