## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
import numpy as np

# 1. Load the starter data
# Note: From work/notebooks/, we need to go up two levels to find the data folder
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 2. Show the unit of analysis (One row = One Content Item/URL)
print(f"Dataset Grain: {df.shape[0]} unique content items.")
print(df[['content_id', 'impressions_90d', 'avg_position', 'days_since_last_update', 'trend_direction']].head(3))

# 3. Sketch the Target Column (Defining the "Decaying Champion")
# A champion is a high-ranker (Pos < 20) in a 'down' trend.
df["is_decaying_champion"] = (
    (df["trend_direction"].str.lower() == "down") & 
    (df["avg_position"] < 20)
).astype(int)

print("\nTarget Column Distribution:")
print(df["is_decaying_champion"].value_counts(normalize=True))
```
